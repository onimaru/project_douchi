```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
```

```python
home_folder = "/home/ninja"
data_path = f"{home_folder}/datasets/titanic"
train_df = pd.read_csv(f"{data_path}/train.csv")
test_df = pd.read_csv(f"{data_path}/test.csv")
gender_df = pd.read_csv(f"{data_path}/gender_submission.csv")
```

```python
test_df = pd.merge(left=test_df,right=gender_df,how="inner",on="PassengerId")[train_df.columns]
```

### Plano

- `PassengerId`: descartar
- `Survived`: rótulo, `y`
- `Pclass`: fica igual
- `Name`: trocar pelas colunas do `bag-of-words`
- `Sex`: transformar em binário
- `Age`: trocar `NaN` pela média
- `SibSp`: fica igual
- `Parch`: fica igual
- `Ticket`: trocar pelas colunas do `bag-of-words`
- `Fare`: transformar em `float`

```python
train_df["Fare"] = train_df["Fare"].astype(float)
train_df["Fare"] = train_df["Fare"].fillna(train_df["Fare"].mean())
train_df["Age"] = train_df["Age"].astype(float)
train_df["Age"] = train_df["Age"].fillna(train_df["Age"].mean())

test_df["Fare"] = test_df["Fare"].astype(float)
test_df["Fare"] = test_df["Fare"].fillna(test_df["Fare"].mean())
test_df["Age"] = test_df["Age"].astype(float)
test_df["Age"] = test_df["Age"].fillna(test_df["Age"].mean())
```

```python
# Ticket
ticket_vectorizer = CountVectorizer(analyzer='word',
                             ngram_range=(1,1),
                             max_df=1.0,
                             min_df=2,
                             max_features=None,
                             vocabulary=None,)

ticket_vectorizer.fit(train_df["Ticket"])
docs = ticket_vectorizer.transform(train_df["Ticket"])

ticket_bow_df = pd.DataFrame(data=docs.toarray(),columns=ticket_vectorizer.get_feature_names_out(),dtype=int)

# feature_ticket = ticket_vectorizer.get_feature_names_out().tolist()
```

```python
# Cabin
train_df["Cabin"] = train_df["Cabin"].fillna("")

cabin_vectorizer = CountVectorizer(analyzer='word',
                             ngram_range=(1,1),
                             max_df=1.0,
                             min_df=1,
                             max_features=None,
                             vocabulary=None,)

cabin_vectorizer.fit(train_df["Cabin"])
docs = cabin_vectorizer.transform(train_df["Cabin"])

cabin_bow_df = pd.DataFrame(data=docs.toarray(),columns=cabin_vectorizer.get_feature_names_out(),dtype=int)

# feature_ticket = ticket_vectorizer.get_feature_names_out().tolist()
```
